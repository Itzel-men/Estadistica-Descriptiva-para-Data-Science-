## Scatter Plot (Diagramas de dispersión)

Se utliza cuando queremos ver las variaciones de una variable con respecto de otra variable, ambas numéricas.
Nos va a dar información cuando queremos trabajar con las **correlaciones**.


```
import pandas as pd
import seaborn as sns
```

Trabajaremos con un dataset llamados `iris` que corresponde a un conjunto de datos que representan atributos especiales de unas flores llamadas iris que vienen tres especies distintas.

```
iris = sns.load_dataset('iris')
iris.head()
```

En seaborn podemos trabajar con scatter plots en la documentación obtenemos más información: [scatter plot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html).

$x$ y $y$ determinan las dimensiones de las variables que queremos contrastar. Cada puntos en el scatter es el atributo en $x$ y $y$ del individuo, es decir, de cada flor.

```
sns.scatterplot(data=iris, x='sepal_length', y='petal_length')
```

Si agregamos una segregación por una variables categórica, en este caso, por especie.

```
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

## Join plot

Nos permite hacer el diagrama de dispersión y en cada eje nos muestra un histograma. Podemos obtener más información en la [documentación](http://seaborn.pydata.org/generated/seaborn.jointplot.html).

```
sns.jointplot(data=iris,x = 'sepal_length',y = 'petal_length')
```

```
sns.jointplot(data=iris,x = 'sepal_length',y = 'petal_length', hue='species')
```

En el caso anterior los histogramas se muestran como distribuciones suaves. Vemos como se separa cada una de las distribuciones.

## Box plot

Podemos encontrar información del box plot en su [documentación](http://seaborn.pydata.org/generated/seaborn.boxplot.html).

Queremos un diagrama de caja por cada una de las especies que tenemos en el dataset y la longitud del sepalo.

```
sns.boxplot(x = 'species', y='sepal_length', data=iris)
```

## Bar plot

Podemos encontrar más información en su [documentación](http://seaborn.pydata.org/generated/seaborn.barplot.html).


```
mean_petal_len = iris.groupby('species')[['petal_length']].mean()
mean_petal_len
```

```
sns.barplot(data = iris, x = 'species',y = 'petal_length')
```
